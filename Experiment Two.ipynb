 Importing libraires of scipy

```python
import numpy as np
from scipy.cluster.vq import kmeans,vq
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
```

Preparing the data set

```python
data = np.vstack((np.random.rand(150,3) + np.array([.5,.5,.5]),np.random.rand(150,3)))
```

Generating the centroids and calculating the distortion present in the dataset for 3 clusters. Unlike the traditional kmeans, this returns distortion as the mean of the eucledian distance between the observations and the centroids

```python
centroids, distortion = kmeans(data,3)
```

```python
code, distortion = vq(data,centroids)
```

```python
print(code)
```

```python
print(centroids)
```

```python
color = {0 : 'r',1 : 'k',2:'g'}
label_color = [color[x] for x in code]
plt.title("K-Means Clustering")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.scatter(data[:,0],data[:,1], c=label_color)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(data[:,0],data[:,1],data[:,2],c=code,s=50)
ax.set_title('K-Means Clustering')
ax.set_xlabel('X label')
ax.set_ylabel('Y label')
ax.set_zlabel('Z label')
plt.colorbar(scatter)
```
